```python
import pandas as pd
from numpy.random import exponential
```

Начнем с того, что мы смотрим как визуализируется график чего бы то ни было в pandas. Сделаем DataFrame, в котором будет две колонки: X и Y. По Y у нас будет экспонециальное распределение

```python
df = pd.DataFrame({'x': range(20), 'y': exponential(10, 20)})
```

```python
%matplotlib inline
```

```python
df.y.hist()
#вызов hist, он изображает гистограмму нашего распределения, где высота столбика показывает количество 
#наблюдений с таким значением, которое показано по Х.
```

Для того чтобы начать работать с библиотекой matplotlib, нам нужно импортировать но мы импортируем не саму библиотеку, а ее модуль Pyplot. Модуль Pyplot позволяет работать с библиотекой Matplotlib в оперативном стиле

```python
import matplotlib.pyplot as plt
```

Самое простое что мы можем сделать - это взять сет, допустим, мы будем изображать успешность игрока в зависимости от номера попытки. У нас есть экспонециальное распределение с параметром Х и у вас будет 20 попыток. И у нас будет второе экспонециальное распределение с параметром 6 и размером 20

```python
data1 = exponential(5, 20)
data2 = exponential(6, 20)
```

```python
plt.plot(data1, label='Первый игрок')
```

Второй метод изображения данных это не непрерывная линия, а точка, где у нас есть кординаты X и Y. Для этого есть и для него должно когда два объекта. В первом мы указываем координаты X, во втором координаты Y. Координаты X - это просто номера попыток, поэтому мы используем метод lh, где удалены количество попыток. А Y - это, собственно, наше распределение. 

```python
plt.scatter(range(len(data2)), data2, color='red', label='Второй игрок')
```

 метод legend и он отобразит легенду. Легенда - это описание того, что изображено на вашем графике и что чем кодируется. Если мы вызовем ее просто так, мы ничего не увидим, просто маленький квадратик, потому что мы нигде не указали что писалось для каждого из графиков. Для этого, при рисовании графика вы можете передать аргумент label. Первый это у нас второй игрок, а здесь label первый игрок.

```python
plt.plot(data1, label='Первый игрок')
plt.scatter(range(len(data2)), data2, color='red', label='Второй игрок')
plt.title('Результаты', fontdict={'fontsize': 20})
plt.xlabel('Номер попытки')
plt.ylabel('Результат')
plt.legend()
plt.xticks(range(0, 20, 4))  # ticks, заметки на оси Х, были только целочисленными
```
